```python
import pretty_midi
import os

notes = []

dataset_path = "maestro-v3.0.0"

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".midi") or file.endswith(".mid"):
            midi = pretty_midi.PrettyMIDI(os.path.join(root, file))

            for instrument in midi.instruments:
                for note in instrument.notes:
                    notes.append(note.pitch)

print("Total notes:", len(notes))
```

```python
import numpy as np
notes1 = notes[:1000000]
sequence_length = 50

X = []
y = []

for i in range(len(notes1) - sequence_length):
    X.append(notes1[i:i+sequence_length])
    y.append(notes1[i+sequence_length])

X = np.array(X)
y = np.array(y)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

model = tf.keras.Sequential([
    layers.Embedding(128, 100, input_length=sequence_length),
    layers.LSTM(256, return_sequences=True),
    layers.LSTM(256),
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy"
)

model.summary()
```

```python
model.fit(
    X,
    y,
    epochs=20,
    batch_size=64
)
```

```python
import random

start = random.randint(0, len(X)-1)
pattern = list(X[start])

generated = []

for i in range(200):
    x = np.array(pattern).reshape(1, sequence_length)

    prediction = model.predict(x, verbose=0)
    note = np.argmax(prediction)

    generated.append(note)
    pattern.append(note)
    pattern = pattern[1:]
```

```python
from music21 import stream, note

output = stream.Stream()

for n in generated:
    output.append(note.Note(int(n)))

output.write("midi", "generated_music.mid")
```

```python
import numpy as np
import random
from tensorflow.keras.models import load_model
from music21 import stream, note, instrument

# -----------------------------
# Load trained emotion-conditioned model
# -----------------------------
model = load_model("music_model_emotion.h5")

# -----------------------------
# Define emotions and vectors
# -----------------------------
emotion_map = {
    "relax": [1, 0, 0, 0],
    "happy": [0, 1, 0, 0],
    "sad":   [0, 0, 1, 0],
    "focus": [0, 0, 0, 1]
}

def get_emotion_vector(user_text):
    """Simple text-based emotion detection. Replace with your NLP model if needed."""
    user_text = user_text.lower()
    if "stress" in user_text or "tired" in user_text:
        return emotion_map["relax"], "relax"
    elif "happy" in user_text or "excited" in user_text:
        return emotion_map["happy"], "happy"
    elif "sad" in user_text or "down" in user_text:
        return emotion_map["sad"], "sad"
    else:
        return emotion_map["focus"], "focus"

# -----------------------------
# Temperature sampling function
# -----------------------------
def sample(predictions, temperature=1.0):
    predictions = np.log(predictions + 1e-9) / temperature
    exp_preds = np.exp(predictions)
    predictions = exp_preds / np.sum(exp_preds)
    return np.random.choice(len(predictions), p=predictions)

# -----------------------------
# Generate music function
# -----------------------------
def generate_music(seed_sequence, user_text, num_notes=200, temperature=1.0):
    emotion_vector, emotion_label = get_emotion_vector(user_text)
    sequence_length = len(seed_sequence)
    pattern = list(seed_sequence)
    generated = []

    # Create MIDI stream
    output = stream.Stream()
    output.append(instrument.Piano())

    for i in range(num_notes):
        x = np.array(pattern + emotion_vector).reshape(1, sequence_length + len(emotion_vector))
        prediction = model.predict(x, verbose=0)[0]
        note_idx = sample(prediction, temperature)
        generated.append(note_idx)

        # -----------------------------
        # Add dynamics (velocity & duration)
        # -----------------------------
        new_note = note.Note(int(note_idx))
        # Velocity
        new_note.volume.velocity = random.randint(50, 100)
        # Duration by emotion
        if emotion_label == "relax":
            new_note.quarterLength = random.choice([1, 2])
        elif emotion_label == "happy":
            new_note.quarterLength = random.choice([0.5, 1])
        elif emotion_label == "sad":
            new_note.quarterLength = random.choice([1, 2, 3])
        elif emotion_label == "focus":
            new_note.quarterLength = 1

        output.append(new_note)

        # Update seed for next note
        pattern.append(note_idx)
        pattern = pattern[1:]

    # Save MIDI
    midi_filename = f"{emotion_label}_music_generated.mid"
    output.write("midi", midi_filename)
    print(f"✅ Music generated: {midi_filename}")

    return generated, emotion_label

# -----------------------------
# Example usage
# -----------------------------
# Replace this with a seed sequence from your training data
seed_sequence = X_new[random.randint(0, len(X_new)-1)][:50]

# Example user input
user_text = "I am feeling very stressed right now"

generated_notes, emotion = generate_music(seed_sequence, user_text, num_notes=200, temperature=1.2)

# -----------------------------
# Optional: print some statistics
# -----------------------------
stats = {
    "min_note": int(np.min(generated_notes)),
    "max_note": int(np.max(generated_notes)),
    "unique_notes": int(len(set(generated_notes))),
    "most_frequent_note": int(np.bincount(generated_notes).argmax())
}
print(f"[{emotion}] Stats → {stats}")
```

```python
import numpy as np
import random
from tensorflow.keras.models import load_model
from music21 import stream, note, instrument

# -----------------------------
# Load model
# -----------------------------
model = load_model("music_model_emotion.h5")

# -----------------------------
# Emotion mapping
# -----------------------------
emotion_map = {
    "relax": [1, 0, 0, 0],
    "happy": [0, 1, 0, 0],
    "sad":   [0, 0, 1, 0],
    "focus": [0, 0, 0, 1]
}

def detect_emotion(user_text):
    user_text = user_text.lower()
    if "stress" in user_text or "tired" in user_text:
        return emotion_map["relax"], "relax"
    elif "happy" in user_text or "excited" in user_text:
        return emotion_map["happy"], "happy"
    elif "sad" in user_text or "down" in user_text:
        return emotion_map["sad"], "sad"
    else:
        return emotion_map["focus"], "focus"

# -----------------------------
# Sampling
# -----------------------------
def sample(predictions, temperature=1.0):
    predictions = np.log(predictions + 1e-9) / temperature
    exp_preds = np.exp(predictions)
    predictions = exp_preds / np.sum(exp_preds)
    return np.random.choice(len(predictions), p=predictions)

# -----------------------------
# Generate music chunk
# -----------------------------
def generate_chunk(seed_seq, emotion_vector, chunk_size=16, temperature=1.0):
    pattern = list(seed_seq)
    generated_notes = []
    chunk_stream = stream.Stream()
    chunk_stream.append(instrument.Piano())

    for _ in range(chunk_size):
        x = np.array(pattern + emotion_vector).reshape(1, len(pattern)+len(emotion_vector))
        prediction = model.predict(x, verbose=0)[0]
        note_idx = sample(prediction, temperature)
        generated_notes.append(note_idx)

        # Dynamics
        velocity = random.randint(50, 100)
        if emotion_vector == emotion_map["relax"]:
            duration = random.choice([1, 2])
        elif emotion_vector == emotion_map["happy"]:
            duration = random.choice([0.5, 1])
        elif emotion_vector == emotion_map["sad"]:
            duration = random.choice([1, 2, 3])
        else:
            duration = 1

        new_note = note.Note(int(note_idx))
        new_note.quarterLength = duration
        new_note.volume.velocity = velocity
        chunk_stream.append(new_note)

        # Update pattern
        pattern.append(note_idx)
        pattern = pattern[1:]

    return pattern, generated_notes, chunk_stream

# -----------------------------
# Adaptive loop (text input)
# -----------------------------
seed_sequence = [60, 62, 64, 65, 67, 69, 71, 72]
full_stream = stream.Stream()
full_stream.append(instrument.Piano())

print("🎵 AI Music Therapy Session (type 'exit' to stop) 🎵")
while True:
    user_text = input("Your emotion: ")
    if user_text.lower() == "exit":
        break

    emotion_vector, emotion_label = detect_emotion(user_text)
    seed_sequence, generated_notes, chunk_stream = generate_chunk(seed_sequence, emotion_vector)
    full_stream.append(chunk_stream)

    # Save incremental MIDI
    midi_filename = f"adaptive_session.mid"
    full_stream.write("midi", midi_filename)

    print(f"[{emotion_label}] Generated {len(generated_notes)} notes → min:{min(generated_notes)} max:{max(generated_notes)} unique:{len(set(generated_notes))}")
    print(f"✅ MIDI saved → {midi_filename}")

print("🎵 Session Ended 🎵")
```
